```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/top-10000-spotify-songs-1960-now/top_10000_1960-now.csv')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Generate basic statistics for numerical columns
numerical_stats = df.describe()

numerical_stats.transpose()
```

```python
# Let's find out the top 10 artists with the most tracks in the dataset
top_artists = df['Artist Name(s)'].value_counts().head(10)

top_artists
```

```python
# Function to safely parse a date and return whether an error occurred
def safe_date_parse(date):
    try:
        pd.to_datetime(date)
        return False  # No error
    except Exception:
        return True  # Error occurred

# Apply the function to the 'Album Release Date' column and find problematic entries
problematic_dates = df['Album Release Date'].apply(safe_date_parse)

# Display the problematic dates
df[problematic_dates]
```

```python
# Remove the problematic row from the DataFrame
df = df[~problematic_dates]

# Verify the removal by checking the shape of the DataFrame
df.shape
```

```python
# Extract the year from the 'Album Release Date' column again
df['Year'] = pd.to_datetime(df['Album Release Date']).dt.year

# Display the first few rows of the DataFrame to verify the new column
df.head()
```

```python
# Group the data by 'Year' and calculate the number of tracks and average popularity for each year
yearly_data = df.groupby('Year').agg({'Track Name': 'count', 'Popularity': 'mean'})

# Rename the columns for clarity
yearly_data.rename(columns={'Track Name': 'Track Count', 'Popularity': 'Average Popularity'}, inplace=True)

# Display the yearly data
yearly_data
```

```python
# Calculate the proportion of explicit tracks
explicit_proportion = df['Explicit'].mean()

# Calculate the average popularity of explicit and non-explicit tracks
average_popularity_by_explicitness = df.groupby('Explicit')['Popularity'].mean()

explicit_proportion, average_popularity_by_explicitness
```

```python
# Calculate the average values of several characteristics for explicit and non-explicit tracks
characteristics = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo']
average_characteristics_by_explicitness = df.groupby('Explicit')[characteristics].mean()

average_characteristics_by_explicitness
```

```python
# Calculate the correlation matrix for a subset of the numerical variables
correlation_matrix = df[characteristics + ['Popularity']].corr()

correlation_matrix
```

```python
from sklearn.model_selection import train_test_split

# Define the features and the target
features = df[characteristics + ['Explicit']]
target = df['Popularity']

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Fill missing values with the mean value of their respective column
X_train_filled = X_train.fillna(X_train.mean())
X_test_filled = X_test.fillna(X_train.mean())  # Note: we use the mean from the training set, not the test set

# Check if there are any missing values left
X_train_filled.isna().sum(), X_test_filled.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a Linear Regression model
model = LinearRegression()

```

```python
# Train the model on the filled training data
model.fit(X_train_filled, y_train)

# Make predictions on the filled test data
y_pred = model.predict(X_test_filled)

# Calculate the root mean squared error (RMSE) and the R-squared score (R2)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

rmse, r2
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the filled training data
rf_model.fit(X_train_filled, y_train)

# Make predictions on the filled test data
y_pred_rf = rf_model.predict(X_test_filled)

# Calculate the root mean squared error (RMSE) and the R-squared score (R2)
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)

rmse_rf, r2_rf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")

# Create a figure and axes for the plots
fig, axs = plt.subplots(2, 2, figsize=(20, 14))

# Plot the top 10 artists with the most tracks
top_artists.plot(kind='barh', ax=axs[0, 0], color='skyblue', edgecolor='black')
axs[0, 0].set_title('Top 10 Artists with the Most Tracks')
axs[0, 0].invert_yaxis()

# Plot the yearly track count
yearly_data['Track Count'].plot(ax=axs[0, 1], color='skyblue')
axs[0, 1].set_title('Yearly Track Count')

# Plot the yearly average popularity
yearly_data['Average Popularity'].plot(ax=axs[1, 0], color='skyblue')
axs[1, 0].set_title('Yearly Average Popularity')

# Plot the correlation matrix heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title('Correlation Matrix Heatmap')

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
# Calculate the average values of song characteristics for each year
average_characteristics_by_year = df.groupby('Year')[characteristics].mean()

average_characteristics_by_year
```
